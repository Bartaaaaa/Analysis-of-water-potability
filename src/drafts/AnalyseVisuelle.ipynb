## 1) Comprendre la donnée :


pH :Mesure de l'acidité ou de l'alcalinité de l'eau sur une échelle de 0 à 14. Un pH de 7 est neutre, les valeurs inférieures à 7 indiquent une acidité, et celles supérieures à 7 une alcalinité.

Dureté : Indique la capacité de l'eau à précipiter le savon, mesurée en milligrammes par litre (mg/L). Elle est principalement déterminée par la concentration des sels de calcium et de magnésium.

Solides : Représente la totalité des solides dissous dans l'eau, mesurée en parties par million (ppm). Cela inclut les minéraux, les sels et les matières organiques.

Chloramines : Mesure la quantité de chloramines présentes dans l'eau, en parties par million (ppm). Les chloramines sont des désinfectants utilisés pour traiter l'eau potable.

Sulfate : Indique la concentration de sulfates dissous dans l'eau, mesurée en milligrammes par litre (mg/L). Les sulfates sont des substances qui peuvent se trouver naturellement dans les minéraux.

Conductivité : Réfère à la conductivité électrique de l'eau, mesurée en microsiemens par centimètre (μS/cm). C'est un indicateur de la quantité d'ions présents dans l'eau.

Carbone Organique : Mesure la quantité de carbone organique trouvée dans l'eau, en parties par million (ppm). Il sert d'indicateur de la qualité de l'eau.

Trihalométhanes : Ce sont des composés chimiques présents dans l'eau, mesurés en microgrammes par litre (μg/L). Ils sont formés comme sous-produit de la chloration.

Turbidité : Mesure de la turbidité ou de la limpidité de l'eau, indiquée en Unités Néphélométriques de Turbidité (NTU). Elle est causée par des particules généralement invisibles à l'œil nu.

Potabilité : C'est un indicateur binaire montrant si l'eau est potable ou non pour la consommation humaine. Une valeur de 1 signifie potable (sûre à boire), et une valeur de 0 signifie non potable (non sûre à boire).

### On va maintenant essayer de voir si la présence forte en quantité d'un élément dans l'eau a un impact sur sa potabilité ou pas.

```python
xtest_ph = df.melt(id_vars=["Potability"], value_vars=["ph"])
xtest_hardness = df.melt(id_vars=["Potability"], value_vars=["Hardness"])
xtest_solids = df.melt(id_vars=["Potability"], value_vars=["Solids"])
xtest_chloramines = df.melt(id_vars=["Potability"], value_vars=["Chloramines"])
xtest_sulfate = df.melt(id_vars=["Potability"], value_vars=["Sulfate"])
xtest_conductivity = df.melt(id_vars=["Potability"], value_vars=["Conductivity"])
xtest_organic_carbon = df.melt(id_vars=["Potability"], value_vars=["Organic_carbon"])
xtest_trihalomethanes = df.melt(id_vars=["Potability"], value_vars=["Trihalomethanes"])
xtest_turbidity = df.melt(id_vars=["Potability"], value_vars=["Turbidity"])

# Création d'une figure avec plusieurs sous-graphes
fig, axes = plt.subplots(3, 3, figsize=(18, 12))  # 3 lignes, 3 colonnes

# Liste des datasets et des titres
datasets = [xtest_ph, xtest_hardness, xtest_solids, xtest_chloramines, xtest_sulfate,
            xtest_conductivity, xtest_organic_carbon, xtest_trihalomethanes, xtest_turbidity]
titles = ["pH", "Hardness", "Solids", "Chloramines", "Sulfate",
          "Conductivity", "Organic Carbon", "Trihalomethanes", "Turbidity"]

# Tracer chaque graphique dans sa case correspondante
for ax, dataset, title in zip(axes.flatten(), datasets, titles):
    sns.boxplot(x="variable", y="value", data=dataset, hue="Potability", ax=ax)
    ax.set_title(f'Distribution of {title} by Potability')
    ax.set_xlabel('')  # Enlever le nom de la variable sur l'axe des x
    ax.set_ylabel('')  # Enlever "value" de l'axe des y
    ax.legend(title='Potability', loc='upper right')  # Ajuster la légende

plt.tight_layout()  # Ajuste automatiquement les sous-graphes
plt.show()
```

Vu que nos données sont fortement décorrélées, on va faire une ACP pour essayer de déterminer une corrélation entre éléments :

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler

# Supposons que 'df' est votre DataFrame déjà chargé avec les données nécessaires.

# Initialisation de l'objet PCA avec 5 composantes
cls = PCA(n_components=5)
mean = df.mean(axis=0)

dfcentered = df - mean
scaler = MinMaxScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
# Ajustement de PCA sur les données et transformation des données

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standardized = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

pcs_df = cls.fit_transform(df_standardized)  # 'df' doit être votre dataset


# Création du graphique à barres pour les variances expliquées par chaque axe
plt.figure(figsize=(12, 6))  # Définit la taille de la fenêtre d'affichage pour les deux graphiques

plt.subplot(1, 2, 1)  # (nombre de lignes, nombre de colonnes, index du graphique)
plt.bar(["Axe 1", "Axe 2", "Axe 3", "Axe 4", "Axe 5"], cls.explained_variance_ratio_)
plt.title("Variances Expliquées par les Axes PCA")
plt.ylabel("Ratio de Variance Expliquée")
plt.xlabel("Axes PCA")

# Second graphique : Scatter plot de PCA
plt.subplot(1, 2, 2)  # Le deuxième graphique sur la même ligne
df2 = pd.DataFrame(pcs_df, columns=[f"PC{i}" for i in range(1, 6)])
sns.scatterplot(x="PC1", y="PC2", hue=df.Potability, data=df2)
plt.title("Scatter Plot des Deux Premiers Composants Principaux")
plt.ylabel("PC2")
plt.xlabel("PC1")

plt.tight_layout()  # Ajuste automatiquement les sous-graphiques pour qu'ils s'intègrent bien dans la figure
plt.show()
```
